```python
for i in survey:
    if i[0] < 5:
        i.append(" < 5 Years")
    elif 5<= i[0] < 10:
        i.append("5-10 Years")
    elif 10<= i[0] < 15:
        i.append("10-15 Years")
    elif 15<= i[0] < 20:
        i.append("15-20 Years")
    elif 20<= i[0] < 25:
        i.append("20-25 Years")
        
    elif i[0] > 25:
        i.append("25+ Years")
```

```python
lessthan5 = []
fiveto10 = []
tento15 = []
fifteento20 = []
twentyto25 = []
moretha25 = []


for i in survey:
    if i[-1] == " < 5 Years":
        lessthan5.append(i[-2])
    elif i[-1] == "5-10 Years":
        fiveto10.append(i[-2])
    elif i[-1] == "10-15 Years":
        tento15.append(i[-2])
    elif i[-1] == "15-20 Years":
        fifteento20.append(i[-2])
    elif i[-1] == "20-25 Years":
        twentyto25.append(i[-2])
    else:
        moretha25.append(i[-2])
```

```python
sum(lessthan5)/len(lessthan5)
```

```python
cat = [lessthan5, fiveto10,tento15, fifteento20,twentyto25, moretha25]

for c in cat:
    print(sum(c)/len(c))
```
