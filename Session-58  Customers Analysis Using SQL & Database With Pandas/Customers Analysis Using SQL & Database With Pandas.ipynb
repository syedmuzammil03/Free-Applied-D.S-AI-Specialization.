```python
import pandas as pd
import sqlite3 as sql
```

```python
conn = sql.connect("stores.db")
```

```python
# customers orders

q = """
    SELECT *  -- Columns
    FROM ORDERS -- Tables
    LIMIT 5  -- No of rows
"""

pd.read_sql(q,conn)
```

```python
# Join Orders and orderdetails tables

q = """
   select * FROM orders o
   INNER JOIN orderdetails od
   ON o.orderNumber = od.orderNumber
   LIMIT 5
   """

pd.read_sql(q,conn)
```

```python
# One column from each table

q = """
   SELECT customerNumber, priceEach FROM orders o
   INNER JOIN orderdetails od
   on od.orderNumber = o.orderNumber
   LIMIT 5
   """

pd.read_sql(q,conn)
```

```python
# Join Product tables

q = """
   SELECT * FROM orders o
   INNER JOIN orderdetails od
   on o.orderNumber = od.orderNumber
   JOIN products p
   on p.productCode = od.productCode
   LIMIT 5
   """

pd.read_sql(q,conn)
```

```python
# One column from each table


q = """
   SELECT customerNumber,priceEach, buyPrice  
   FROM orders o
   INNER JOIN orderdetails od
   on o.orderNumber = od.orderNumber
   JOIN products p
   on p.productCode = od.productCode
   LIMIT 5
   """

pd.read_sql(q,conn)
```

```python
# Customer Profit Margin


q = """
   SELECT customerNumber,(priceEach -buyPrice) profit_margin  
   FROM orders o
   INNER JOIN orderdetails od
   on o.orderNumber = od.orderNumber
   JOIN products p
   on p.productCode = od.productCode
   LIMIT 5
   """

pd.read_sql(q,conn)
```

```python
# add column quantityOrdered


q = """
   SELECT customerNumber,(priceEach -buyPrice) profit_margin, quantityOrdered
   FROM orders o
   INNER JOIN orderdetails od
   on o.orderNumber = od.orderNumber
   JOIN products p
   on p.productCode = od.productCode
   LIMIT 5
   """

pd.read_sql(q,conn)
```

```python
# profit calculation


q = """
   SELECT customerNumber,(priceEach -buyPrice)*quantityOrdered profit
   FROM orders o
   INNER JOIN orderdetails od
   on o.orderNumber = od.orderNumber
   JOIN products p
   on p.productCode = od.productCode
   """

pd.read_sql(q,conn)
```

```python
# total profit by customer -> Group BY

q = """
   SELECT customerNumber,sum((priceEach -buyPrice)*quantityOrdered) total_profit
   FROM orders o
   INNER JOIN orderdetails od
   on o.orderNumber = od.orderNumber
   JOIN products p
   on p.productCode = od.productCode
   GROUP BY 1
   """

pd.read_sql(q,conn)
```

```python
# storing total profit by customers in a temporary table -> WITH Clause


q = """
    WITH
    profit_by_customers AS
                      (
                       SELECT customerNumber,sum((priceEach -buyPrice)*quantityOrdered) total_profit
                       FROM orders o
                       INNER JOIN orderdetails od
                       on o.orderNumber = od.orderNumber
                       JOIN products p
                       on p.productCode = od.productCode
                       GROUP BY 1
                       )
                       
    SELECT * FROM profit_by_customers
   """

pd.read_sql(q,conn)
```

```python
# Join profit_by_customer (temporary table)


q = """
    WITH
    profit_by_customers AS
                      (
                       SELECT customerNumber,sum((priceEach -buyPrice)*quantityOrdered) total_profit
                       FROM orders o
                       INNER JOIN orderdetails od
                       on o.orderNumber = od.orderNumber
                       JOIN products p
                       on p.productCode = od.productCode
                       GROUP BY 1
                       )
                       
    SELECT * FROM customers c
    JOIN profit_by_customers pc
    ON pc.customerNumber = c.customerNumber
   """

pd.read_sql(q,conn)
```

```python
# Customer Data


q = """
    WITH
    profit_by_customers AS
                      (
                       SELECT customerNumber,sum((priceEach -buyPrice)*quantityOrdered) total_profit
                       FROM orders o
                       INNER JOIN orderdetails od
                       on o.orderNumber = od.orderNumber
                       JOIN products p
                       on p.productCode = od.productCode
                       GROUP BY 1
                       )
                       
    SELECT contactLastName,contactFirstName, city, country, total_profit
    FROM customers c
    JOIN profit_by_customers pc
    ON pc.customerNumber = c.customerNumber
   """

pd.read_sql(q,conn)
```

```python
# Top 5 VIP customers: Bring in the most profit for the store.



q = """
    WITH
    profit_by_customers AS
                      (
                       SELECT customerNumber,sum((priceEach -buyPrice)*quantityOrdered) total_profit
                       FROM orders o
                       INNER JOIN orderdetails od
                       on o.orderNumber = od.orderNumber
                       JOIN products p
                       on p.productCode = od.productCode
                       GROUP BY 1
                       )
                       
    SELECT contactLastName,contactFirstName, city, country, total_profit
    FROM customers c
    JOIN profit_by_customers pc
    ON pc.customerNumber = c.customerNumber
    ORDER BY total_profit DESC
    LIMIT 5
   """

pd.read_sql(q,conn)
```

```python
# Top 5 less engaging customers

q = """
    WITH
    profit_by_customers AS
                      (
                       SELECT customerNumber,sum((priceEach -buyPrice)*quantityOrdered) total_profit
                       FROM orders o
                       INNER JOIN orderdetails od
                       on o.orderNumber = od.orderNumber
                       JOIN products p
                       on p.productCode = od.productCode
                       GROUP BY 1
                       )
                       
    SELECT contactLastName,contactFirstName, city, country, total_profit
    FROM customers c
    JOIN profit_by_customers pc
    ON pc.customerNumber = c.customerNumber
    order by total_profit 
    LIMIT 5
   """

pd.read_sql(q,conn)
```
